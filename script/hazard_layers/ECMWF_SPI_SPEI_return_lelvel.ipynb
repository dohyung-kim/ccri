```python
# === Full Gumbel-based SPEI/SPI raster return level processor ===

import os
import numpy as np
import rasterio
from rasterio.windows import Window
from osgeo import gdal
from tqdm.contrib.concurrent import process_map

# === Parameters ===
input_folder = "spi12_tifs"  # folder with annual SPEI GeoTIFFs
output_folder = "spi12_return_level"  # folder to save return level GeoTIFFs
os.makedirs(output_folder, exist_ok=True)
years = list(range(1940, 2025))  # adapt if needed
return_periods = [10, 30, 50, 100]
vrt_path = os.path.join(input_folder, "spi12_annual_min.vrt")
tif_template = os.path.join(input_folder, "spi12_min_{}.tif")

# === Step 1: Build VRT stack ===
tif_paths = [tif_template.format(year) for year in years]
if not os.path.exists(vrt_path):
    print("🛠️ Building VRT stack...")
    gdal.BuildVRT(vrt_path, tif_paths, separate=True)
else:
    print("✅ VRT stack already exists.")

# === Step 2: Open VRT and get metadata ===
with rasterio.open(vrt_path) as src:
    height, width = src.height, src.width
    count = src.count
    crs = src.crs
    transform = src.transform
    dtype = rasterio.float32
    blockxsize, blockysize = 256, 256

    profile = {
        'driver': 'GTiff',
        'height': height,
        'width': width,
        'count': 1,
        'dtype': dtype,
        'crs': crs,
        'transform': transform,
        'tiled': True,
        'blockxsize': blockxsize,
        'blockysize': blockysize,
        'compress': 'lzw'
    }

    all_windows = [
        Window(col, row, min(blockxsize, width - col), min(blockysize, height - row))
        for row in range(0, height, blockysize)
        for col in range(0, width, blockxsize)
    ]

# === Gumbel fitting for SPEI/SPI (minima) ===
def fit_gumbel_tile_spei(data, return_periods=[10, 30, 50, 100]):
    """
    Fit Gumbel distribution to *minima* of SPEI/SPI values (negative drought extremes).
    Assumes input shape: (years, height, width)
    """
    years, h, w = data.shape
    flat = data.reshape(years, -1)  # (years, pixels)

    valid_counts = (~np.isnan(flat)).sum(axis=0)
    enough_data = valid_counts > 3
    valid_mask = enough_data

    result = np.full((flat.shape[1], len(return_periods)), np.nan, dtype=np.float32)

    if np.any(valid_mask):
        valid_data = flat[:, valid_mask]
        flipped = -1 * valid_data  # because Gumbel is for maxima, and we want minima

        gamma = 0.5772
        mean = np.nanmean(flipped, axis=0)
        std = np.nanstd(flipped, axis=0)

        loc = mean - gamma * std
        scale = std * np.sqrt(6) / np.pi

        for i, rp in enumerate(return_periods):
            result[valid_mask, i] = -1 * (loc - scale * np.log(-np.log(1 - 1 / rp)))

    return result.reshape(h, w, len(return_periods))

# === Tile processor ===
def process_tile(win):
    with rasterio.open(vrt_path) as src:
        data = src.read(window=win)  # shape: (years, h, w)
    result = fit_gumbel_tile_spei(data, return_periods)
    return (win, result)

# === Step 3: Process all tiles in parallel ===
print(f"🚀 Processing {len(all_windows)} tiles with Gumbel fitting...")
results = process_map(process_tile, all_windows, max_workers=os.cpu_count(), chunksize=1, desc="Processing Tiles")

# === Step 4: Write output raster(s) ===
print("💾 Writing return level GeoTIFFs...")
with rasterio.open(vrt_path) as src:
    profile['dtype'] = rasterio.float32
    for i, rp in enumerate(return_periods):
        out_path = os.path.join(input_folder, f"SPI12_return_level_{rp}yr.tif")
        with rasterio.open(out_path, 'w', **profile) as dst:
            for win, result in results:
                dst.write(result[:, :, i], 1, window=win)

print("\n✅ SPEI/SPI Gumbel-based return level maps saved!")
```

```python
# Full code to extract pixel series, fit Gumbel and GEV, and compare return levels safely

import os
import numpy as np
import rasterio
from rasterio.windows import Window
from scipy.stats import genextreme
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import genpareto

# === Parameters ===
vrt_path = "/content/spi12_tifs/spi12_annual_min.vrt"
target_x, target_y = 74,40.7  # longitude, latitude
return_periods = [10, 30, 50, 100]

# === Step 1: Extract pixel series safely ===
with rasterio.open(vrt_path) as src:
    row, col = src.index(target_x, target_y)
    pixel_series = [src.read(i, window=Window(col, row, 1, 1))[0, 0] for i in range(1, src.count + 1)]

pixel_series = np.array(pixel_series, dtype=np.float32)
pixel_series = pixel_series[~np.isnan(pixel_series)]

flipped = -1 * pixel_series  # because Gumbel is for maxima, and we want minima

gamma = 0.5772
mean = np.nanmean(flipped, axis=0)
std = np.nanstd(flipped, axis=0)

loc = mean - gamma * std
scale = std * np.sqrt(6) / np.pi

gumbel_l_return_levels = [
    -1 * (loc - scale * np.log(-np.log(1 - 1 / rp)))
    for rp in return_periods
]
```

```python
from scipy.stats import genextreme
import numpy as np
import pandas as pd

# Your SPEI annual minimum data
series = pixel_series

# Fit using GEV (with convention for minima: fit negative of series and flip result)
shape, loc, scale = genextreme.fit(-series)  # fitting to maxima of -series == minima

# Compute return levels (and flip sign back)
return_periods = [10, 30, 50, 100]
return_levels = [-genextreme.ppf(1 - 1 / rp, shape, loc=loc, scale=scale) for rp in return_periods]

# Present results
df = pd.DataFrame({
    "Return Period (years)": return_periods,
    "Return Level (GEV fitted to minima)": np.round(return_levels, 3)
})

df
```
